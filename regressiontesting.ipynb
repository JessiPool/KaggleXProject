```python
# Load dataset for regression model - US Rental Listings, 2014-2020
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data/KaggleX/pythondata/usrent_merge.csv')
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df.set_index('year_only')['percentage'] \
.plot(style = '.', title = 'Affordability Percentage by Year')
plt.show()
```

```python
# Building the regression model - US Rental Listings, 2014-2020
```

```python
X = df.drop(columns = 'percentage')
```

```python
y = df['percentage']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
c = lr.intercept_
```

```python
c
```

```python
m = lr.coef_
```

```python
y_pred_train = lr.predict(X_train)
```

```python
y_pred_test = lr.predict(X_test)
```

```python
y_pred_train
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train, y_pred_train)
plt.xlabel("Actual Affordability Percentage")
plt.ylabel("Predicted Affordability Percentage")
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_pred_train)
```

```python
y_pred_test = lr.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred_test)
plt.xlabel("Actual Affordability Percentage")
plt.ylabel("Predicted Affordability Percentage")
plt.show()
```

```python
r2_score(y_test, y_pred_test)
```
